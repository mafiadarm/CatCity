```python
import time
from io import BytesIO
from PIL import ImageGrab, Image

import pyautogui as pt
import ddddocr

class ImageOrc:
    def __init__(self):       
        self.ocr = ddddocr.DdddOcr()
    
    def trans_orc(self, image):
        img_byte = BytesIO()
        image.save(img_byte, format='JPEG')
        binary_content = img_byte.getvalue()
        try:
            res = self.ocr.classification(binary_content)
            return res
        except:
            print('臣妾做不到啊')

orc = ImageOrc()
```

```python
# 资源池
def into_pool(pos):
    while 1:
        pt.click(pos)
        time.sleep(1)
        im2 = ImageGrab.grab(bbox=(863, 517, 954, 549))
        t2 = orc.trans_orc(im2)
        if t2 == '开始调查':
            break
            
    pt.click(906, 531)
    
    #确认不用一键上阵(默认有前一次队伍) 
    
    # 确认开自动
    while 1:
        time.sleep(1)
        if pt.pixel(862, 486) == (0, 181, 189):
            break
        else:
            pt.click(888, 478)
    
    # 开三倍   
    while 1:
        time.sleep(1)
        if pt.pixel(734, 566) == (0, 221, 230):
            break
        else:
            pt.click(734, 566)
    
    # 点击确认,确保进入
    while 1:
        pt.click(924, 562)
        im3 = ImageGrab.grab(bbox=(946, 549, 993, 575))
        t3 = orc.trans_orc(im3)
        if t3 != '确认':
            break
    
    # 等待退出
    while 1:
        pt.click(565, 555)  # 防升级遮住
        time.sleep(5)
        im4 = ImageGrab.grab(bbox=(932, 528, 981, 556))
        t4 = orc.trans_orc(im4)
        if t4 == '退出':
            while 1:
                pt.click(956, 539)
                im4 = ImageGrab.grab(bbox=(932, 528, 981, 556))
                t4 = orc.trans_orc(im4)
                if t4 != '退出':
                    break
            break
    
    # 退出到地图界面
    while 1:
        pt.click(118, 67)
        time.sleep(2)
        im5 = ImageGrab.grab(bbox=(561, 270, 611, 302))
        t5 = orc.trans_orc(im5)
        if t5 == '活动':
            break            
```

```python
# 涂鸦收集
kaishidiaocha = (911, 520, 1046, 560)  # 开始调查
yijianshangzhen = (975, 460, 1060, 488)  # 一键上阵
fangqi = (939, 534, 1055, 566)  # 放弃
tuichu = (917, 522, 983, 557)  # 退出

def in_start():
    '''
        从开始界面能进入到地图页面
    '''
    im1 = ImageGrab.grab(bbox=kaishidiaocha)  # 左上角 右下角 
    t1 = orc.trans_orc(im1)

    if t1 == '开始调查':
        pt.moveTo(982, 538)  # 点击‘开始调查’
        pt.click()
        
        time.sleep(1)

        while 1:
            im0 = ImageGrab.grab(bbox=fangqi)  # 截取‘放弃’位置图片
            t0 = orc.trans_orc(im0)
            
            if t0 == '放弃':
                break
                
            pt.moveTo(982, 538)  # 如果没点进去 重新点击‘开始调查’
            pt.click()  
            
def in_end_war():
    '''
        进入最后的战斗
    '''
    pt.moveTo(590, 300)  # 点击屏幕中间
    pt.click()
        
    time.sleep(1)

    while 1:   
        im2 = ImageGrab.grab(bbox=yijianshangzhen)  # 检查一键上阵按钮 
        t2 = orc.trans_orc(im2)

        if t2 == '一键上阵':
            break

        pt.moveTo(590, 300)  # 如果没点进去，重新点击屏幕中间
        pt.click()

        
def start_war():
    '''
        进入战斗界面，开始战斗
    '''
    while 1:
        pt.moveTo(1019, 477)  # 点击‘一键上阵’
        pt.click()

        time.sleep(1)

        pt.moveTo(966, 560)  #  点击‘开始’
        pt.click()
        
        time.sleep(1)

        im2 = ImageGrab.grab(bbox=yijianshangzhen)  # 再次检查一键上阵按钮
        t2 = orc.trans_orc(im2)

        if t2 != '一键上阵':  # 一键上阵按钮 不存在说明游戏开始了
            break

            
def exit_click():
    '''
        识别、点击退出按钮
    '''
    im3 = ImageGrab.grab(bbox=tuichu)  # 左上角 右下角 
    t3 = orc.trans_orc(im3)

    if t3 == '退出':
        time.sleep(3)
        pt.moveTo(951, 536)  # 点击‘退出’
        pt.click()
        

def ensure_station_exit():
    while 1:
        exit_click()  # 点击退出
        
        im0 = ImageGrab.grab(bbox=fangqi)  # 截取‘放弃’位置图片
        t0 = orc.trans_orc(im0)

        if t0 == '放弃':
            break
            
        time.sleep(5)

while 1:
    pt.click(855, 441)  # 点击进入通道
    im5 = ImageGrab.grab(bbox=(561, 270, 611, 302))
    t5 = orc.trans_orc(im5)
    if t5 != '活动':
        break  
        
time.sleep(1)
for _ in range(5):
    pt.click(478, 216)  # 选🐖
time.sleep(1)
for _ in range(5):
    pt.click(871, 228)  # 选难度8
time.sleep(1)

# 开三倍   
while 1:
    time.sleep(1)
    if pt.pixel(799, 545) == (0, 255, 243):
        break
    else:
        pt.click(799, 545)

im_typ = ImageGrab.grab(bbox=(963, 41, 1043, 65))
t5_typ = orc.trans_orc(im_typ)

ox = int(t5_typ) // 120 # 剩余票 整除 消费票数

for _ in range(ox):
    in_start()  # 进入到地图界面
    time.sleep(1)
    
    #-----------拿支援点位--8级点--------
    pt.moveTo(822, 323)  # 点位1
    pt.click()
    start_war()  # 开始战斗
    time.sleep(30)
    ensure_station_exit()  # 退出到地图界面
    time.sleep(2)
    
    pt.moveTo(818, 439)  # 点位2
    pt.click() 
    start_war()  # 开始战斗
    time.sleep(30)
    ensure_station_exit()  # 退出到地图界面
    time.sleep(2)
    #-----------拿支援点位--8级点--------
    
    in_end_war()  # 进入到终战
    time.sleep(1)
    start_war()  # 开始战斗
    time.sleep(60)
    
    while 1:  # 退出机制
        pt.moveTo(582, 532)  # 点击空白处 防升级遮住
        pt.click()  
        
        time.sleep(2)
        
        exit_click()  # 退出按钮

        # 确认退出成功
        im1 = ImageGrab.grab(bbox=kaishidiaocha)  # 左上角 右下角 
        t1 = orc.trans_orc(im1)

        if t1 == '开始调查':
            break
        
        # 如果失败，会退回到全局页面
        im0 = ImageGrab.grab(bbox=fangqi)  # 截取‘放弃’位置图片
        t0 = orc.trans_orc(im0)
        
        if t0 == '放弃':  # 重新打
            in_end_war() # 进入到终战
            time.sleep(1)
            start_war()  # 开始战斗
            time.sleep(60)
            
        time.sleep(5)        

# 退出到地图界面
while 1:
    pt.click(118, 67)
    time.sleep(2)
    im5 = ImageGrab.grab(bbox=(561, 270, 611, 302))
    t5 = orc.trans_orc(im5)
    if t5 == '活动':
        break           
```
